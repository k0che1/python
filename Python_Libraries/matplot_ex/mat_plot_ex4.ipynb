```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

def f(x):
  return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
def g(x):
  return 1-x**3+x**5
def h(x):
  return (5 * (np.sin(np.exp(0.8*x)))**2 * x**2) / np.exp(x**2 / 2)
```

```python
#Задание 1
x = np.linspace(-20, 100, 100)
fig, axes = plt.subplots(1,3, figsize=(13,6))

axes[0].plot(x,f(x),'r',label = '$f(x)= \\sin(\\frac{x}{5})\\cdot e^\\frac{x}{10}+5\\cdot e^\\frac{-x}{2}$')
axes[0].plot(x,g(x),'g',label = '$g(x)= 1-x^3+x^5$')
axes[0].plot(x,h(x),'b',label = '$h(x)=\\frac{5\\cdot \\sin ^2(e^0.8x)\\cdot x^2}{e^\\frac{x^2}{2}}$')
axes[0].legend(fontsize=10)
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')
axes[0].set_ylim(-200,200)

x = np.linspace(-2, 4, 100)
axes[1].plot(x,f(x),'r',label = '$f(x)= \\sin(\\frac{x}{5})\\cdot e^\\frac{x}{10}+5\\cdot e^\\frac{-x}{2}$')
axes[1].plot(x,g(x),'g',label = '$g(x)= 1-x^3+x^5$')
axes[1].plot(x,h(x),'b',label = '$h(x)=\\frac{5\\cdot \\sin ^2(e^0.8x)\\cdot x^2}{e^\\frac{x^2}{2}}$')
axes[1].legend(fontsize=10)
axes[1].set_xlabel('x')
axes[1].set_ylabel('f(x)')
axes[1].set_ylim(-50,20)

x = np.linspace(-5,15, 100)
axes[2].plot(x,f(x),'r',label = '$f(x)= \\sin(\\frac{x}{5})\\cdot e^\\frac{x}{10}+5\\cdot e^\\frac{-x}{2}$')
axes[2].plot(x,g(x),'g',label = '$g(x)= 1-x^3+x^5$')
axes[2].plot(x,h(x),'b',label = '$h(x)=\\frac{5\\cdot \\sin ^2(e^0.8x)\\cdot x^2}{e^\\frac{x^2}{2}}$')
axes[2].legend(fontsize=10)
axes[2].set_xlabel('x')
axes[2].set_ylabel('f(x)')
axes[2].set_ylim(-2,4)


plt.suptitle('Исследование функций')

fig.tight_layout()
plt.show()



```

Проанализируем каждую функцию:

**f(x) = sin(x/5) * exp(x/10) + 5 * exp(-x/2) **

Содержит синусоидальные колебания, умноженные на растущую экспоненту (exp(x/10)), что вызывает увеличение амплитуды при больших x.
Вторая часть — убывающая экспонента (exp(-x/2)), которая стремится к нулю при больших x.
В целом, функция сочетает колебания с экспоненциальным ростом и убыванием, что приводит к сложному поведению: при больших x доминирует экспоненциальный рост, а при малых — колебания.

Приблизительные экстремумы на интервале (0,6):

Первый локальный минимум примерно около x≈1.2.
Второй локальный максимум примерно около x≈3.0.
Второй локальный минимум около x≈4.8.

Точка перегиба примерно в x = 1.

**g(x) = 1 - x^3 + x^5**

Полиномиальная функция пятой степени с доминирующим членом x^5.
При больших |x| функция растет очень быстро, особенно при больших x.
В окрестности нуля — плавный переход, возможны экстремумы.
Характер: полиномиальная, быстро растущая при больших |x|, с возможными экстремумами.

Экстремумы на интервале (−1,2):
Максимум около x≈−0.775.
Минимум около x≈0.775

В точке ( x = -(3.10)^0.5 ) происходит смена знака ( g''(x) ) с отрицательного на положительный — точка перегиба.
В точке ( x = 0 ) — смена знака с положительного на отрицательный — точка перегиба.
В точке ( x = (3.10)^0.5 ) — смена знака с отрицательного на положительный — точка перегиба.

**h(x) = (5 * (sin^2(exp^(0.8 * x))) * x^2) / exp((x^2)/2)**

В числителе: синус в квадрате, умноженный на x^2 и на 5.
В знаменателе: экспонента с x^2/2, которая растет очень быстро при больших |x|.
Поскольку синус в квадрате ограничен 0 и 1, числитель ограничен и растет медленно, а знаменатель экспоненциально увеличивается.
В результате, при больших |x| функция стремится к нулю.
Характер: убывающая функция, стремящаяся к нулю при больших |x|.

Экстремумы на интервале  (0,3):

x≈0.56;
x≈1.94.

Точка перегиба на интервале (0,2):
примерно в x = 0.8, x = 1.4, x = 1.8

Все три функции апериодичны.
