```python
import numpy as np
import matplotlib.pyplot as plt


with open ('1.png','rb') as osn:
  osn_img=plt.imread(osn)

with open ('2.png','rb') as bot1:
  bot1_img=plt.imread(bot1)

with open ('3.png','rb') as bot2:
  bot2_img=plt.imread(bot2)

with open ('4.png','rb') as bot3:
  bot3_img=plt.imread(bot3)

fig = plt.figure(figsize=(7,5))
spec = fig.add_gridspec(2, 3, left=0.01, right=0.5,wspace=0)

ax0 = fig.add_subplot(spec[0, :])
ax0.imshow(osn_img)
ax0.get_yaxis().set_visible(False)
ax0.get_xaxis().set_visible(False)

ax10 = fig.add_subplot(spec[1, 0])
ax10.imshow(bot1_img)
ax10.get_yaxis().set_visible(False)
ax10.get_xaxis().set_visible(False)

ax11 = fig.add_subplot(spec[1, 1])
ax11.imshow(bot2_img)
ax11.get_yaxis().set_visible(False)
ax11.get_xaxis().set_visible(False)

ax12 = fig.add_subplot(spec[1, 2])
ax12.imshow(bot3_img)
ax12.get_yaxis().set_visible(False)
ax12.get_xaxis().set_visible(False)
plt.show()
```
