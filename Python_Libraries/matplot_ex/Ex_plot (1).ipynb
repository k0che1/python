```python
#Заданиe 1 Часть 1
import pandas as pd
import numpy as np
import  matplotlib.pyplot  as  plt

def f(x,L,B):
    return (x**B+L**B)/(x**B)

x = np.arange(-5, 5, 0.1)
x[np.where(np.abs(x)<0.1)]=np.nan

plt.figure(figsize=(10,5))
plt.plot(x,f(x,1,1),'r',label = u'α=1, β=1', linewidth=0.5)
plt.plot(x,f(x,2,1),'g',label = u'α=2, β=1', linewidth=0.5)
plt.plot(x,f(x,1,2),'b',label = u'α=1, β=2', linewidth=0.5)
plt.ylim(-20, 30)
plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.legend()
plt.title('График функции $f(x)=\\frac{x^β + α^β​}{x^β}$ при различных  α и β')
plt.savefig('1.svg')
```

Поведение зависит от знака α: если α>0, функция убывает от бесконечности  к 1; если α<0, функция убывает, оставаясь  ниже 1, и может иметь область отрицательных значений.

```python
#Заданиe 1 Часть 3
import numpy as np
import matplotlib.pyplot as plt

def f(x,L,B):
    return (x**B+L**B)/(x**B)

x = np.arange(-5, 0, 0.1)

fig=plt.figure(figsize=(9,6))
axes1=fig.add_axes([0.1, 0.1, 1, 1])
axes1.plot(x,f(x,1,1),'r',label = u'α=1, β=1')
axes1.plot(x,f(x,2,1),'g',label = u'α=2, β=1')
axes1.plot(x,f(x,1,2),'b',label = u'α=1, β=2')
axes1.plot(x,x*0,'m',label = u'f(x)=0')
axes1.legend()
plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.title('График функции $f(x)=\\frac{x^β + α^β​}{x^β}$ при различных  α и β')

x = np.arange(-10000, -1000, 0.1)
axes2=fig.add_axes([0.4, 0.5, 0.5, 0.4])
axes2.plot(x,f(x,1,1),'r',label = u'α=1, β=1')
axes2.plot(x,f(x,2,1),'g',label = u'α=2, β=1')
axes2.plot(x,f(x,1,2),'b',label = u'α=1, β=2')


axes2.legend()

plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.title('График функции $f(x)=\\frac{x^β + α^β​}{x^β}$ при удалении x от 0 к −∞')
plt.savefig('3.svg')
```

Вывод:  При x→−∞, независимо от знака α^β и четности β, функция стремится к 1.
Поведение ближе к x→−∞: если β — нечетное, x^β→ − ∞, f(x)→1;
если β — четное, x^β→−∞, f(x)→1.

При β>:

f(x) убывает на [0,+∞); Стремится к ∞ при x→0, к 1 при x→+∞.

При β<0:

f(x) может возрастать или убывать в зависимости от знака α^β; Стремится к ∞ (если α^β>0), к 1 — при x→+∞.

Пересечения графиков:

— возможны при определённых соотношениях α и β.

— в целом, графики могут пересекаться, если значения функций для разных параметров совпадают при некоторых x.

Общие тенденции:

— графики стремятся к горизонтальной асимптоте f(x)→1 при больших x (при β>0 и β<0).

— Поведение функции вблизи нуля зависит от знака и величины α,β.

```python
#Заданиe 2
import pandas as pd
import numpy as np
import  matplotlib.pyplot  as  plt

def f(x,L,B):
    return (x**B+L**B)/(x**B)

x = np.arange(-5, 5, 0.1)
#x[np.where(np.abs(x)<0.1)]=np.nan

plt.figure(figsize=(13,8))
plt.plot(x,f(x,1,0),label = u'α=1, β=0')
plt.plot(x,f(x,0,0), linestyle='--',label = u'α=0, β=0')
plt.plot(x,f(x,0,1),label = u'α=0, β=1')
plt.plot(x,f(x,0,2), linestyle='--',label = u'α=0, β=2')
plt.plot(x,f(x,1,-1),label = u'α=1, β=-1')
plt.plot(x,f(x,2,-1),label = u'α=2, β=-1')

plt.xlabel(u'x')
plt.ylabel(u'f(x)')
plt.grid()
plt.legend()
plt.title('График функции $f(x)=\\frac{x^β + α^β​}{x^β}$ при различных  α и β')
plt.savefig('4.svg')
```

Вывод:

Рассмотрим каждый случай:

(α=1, β=0): Функция постоянна и равна 2.

(α=0, β=0):Постоянна и равна 2.

(α=0, β=1): Функция равна 1 для всех x ≠ 0.

(α=0, β=2): Также равна 1.

(α=1, β=-1): Линейная функция, возрастает при x > 0.

(α=2, β=-1):Линейная, возрастает при x > 0.


Общий вывод:

При (β=0) функция постоянна и равна 2.

При (α=0) и (β>0) функция равна 1.

При (β<0) и (α ≠ 0) функция — линейная в x, с наклоном, зависящим от α и β.


```python
#Задание 3
import matplotlib.pyplot as plt
import numpy as np

def f(x,L,B):
    return (np.power(x, B) + np.power(L, B)) / np.power(x, B)

x = np.arange(0.1, 5, 0.1)

plt.figure(figsize=(15,8))
plt.plot(x,f(x,1,0.5),label = u'α=1, β=0.5')
plt.plot(x,f(x,1,-0.5), label = u'α=1, β=-0.5')
plt.plot(x,f(x,1,-1.5),label = u'α=1, β=-1.5')

plt.xlabel(u'x')
plt.ylabel(u'f(x)')
#plt.ylim(-1,6)
plt.legend()
plt.grid()
plt.title('График функции $f(x)=\\frac{x^β + α^β​}{x^β}$ при различных  α и β')
plt.savefig('5.svg')
```

Вывод:

**График функции, при α=1, β=0.5:**

Так как для (x <= 0) выражение x^0.5 (квадратный корень из x) не определено в области действительных чисел. Поэтому, чтобы построить график, будем рассматривать только (x > 0).
График показывает функцию, которая стремится к бесконечности при приближении к (x=0).

При увеличении x, значение функции стремится к 1.

В области (x ∈ (0, 5]) функция убывает и приближается к 1 .
График представляет собой гиперболическую кривую, убывающую с ростом (x).

**График функции, при α=1, β=-0.5**:

Для определения sqrt(x) необходимо, чтобы (x >= 0). Поэтому диапазон x лучше взять в ([0, 5]).

График функции показывает убывающую кривую, начиная с (f(0) = 1) и убывая по мере увеличения x.

**График функции, при α=1, β=-1.5:**

Для отрицательных x, (x^-1.5) не определено в области действительных чисел, так как (x^-1.5) = (x*sqrt(x)), а sqrt(x) не определено для отрицательных x.Следовательно, график существует только для (x >= 0).
График функции  показывает возрастающую кривую, начиная с f(0) = 1.

**Общий вывод:**

При (β > 0), функция выглядит как (f(x) = 1 + (α^β)/(x^β)). Так как (x^β) возрастает с x, f(x) убывает при увеличении x.

При (β < 0), (x^β) убывает с ростом x, и функция будет вести себя иначе.

(α=1, β=0.5):
(f(x) = 1 + (α^β)/(x^β)).
Так как x^0.5 растет с ростом x, (f(x)) убывает, и будет выпуклой вниз (вогнутой).

(α=1, β=-0.5): x^-0.5 убывает с ростом x, значит f(x) возрастает и будет выпуклой вверх.

(α=1, β=-1.5):  x^-1.5 возрастает, и поведение функции схоже с предыдущим, но с более выраженной возрастающей тенденцией и будет выпуклой вниз.

При заданных параметрах α и β, графики функции пересекаются в точке (1,2)

```python
#Задание 4
import numpy as np
import matplotlib.pyplot as plt

def f(x,L,B):
    return (x**B+L**B)/(x**B)

x1 = np.arange(-2, 5, 0.1)
x2 = np.arange(0.1, 5, 0.1)

fig, axes = plt.subplots(1,3, figsize=(13,6))

# Plot 1
axes[0].plot(x1,f(x1,1,0), 'b--',label = u'α=1, β=0')
axes[0].plot(x1,f(x1,1,-1),'r--',label = u'α=1, β=-1')
axes[0].plot(x2,f(x2,1,0.5),label = u'α=1, β=0.5')
axes[0].plot(x2,f(x2,1,0.8),label = u'α=1, β=0.8')
axes[0].legend()
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')

# Plot 2
axes[1].plot(x1,f(x1,1,0), 'b--',label = u'α=1, β=0')
axes[1].plot(x1,f(x1,1,-1),'r--',label = u'α=1, β=-1')
axes[1].plot(x2,f(x2,1,-0.5),label = u'α=1, β=-0.5')
axes[1].plot(x2,f(x2,1,-0.8),label = u'α=1, β=-0.8')
axes[1].legend()
axes[1].set_xlabel('x')
axes[1].set_ylabel('f(x)')

#Plot 3
axes[2].plot(x1,f(x1,1,0), 'b--',label = u'α=1, β=0')
axes[2].plot(x1,f(x1,1,-1),'r--',label = u'α=1, β=-1')
axes[2].plot(x2,f(x2,1,-1.5),label = u'α=1, β=-1.5')
axes[2].plot(x2,f(x2,1,-2.5),label = u'α=1, β=-2.5')
axes[2].legend()
axes[2].set_xlabel('x')
axes[2].set_ylabel('f(x)')


plt.suptitle('График функции $f(x)=\\frac{x^β + α^β​}{x^β}$ при различных  α и β')
fig.tight_layout()
plt.savefig('6.svg')
```

Вывод: Все графики проходят через общую точку (1, 2).
Видно, что изменение β значительно влияет на форму графика функции. При отрицательных значениях β функция имеет особенности вблизи x = 0 и степенной рост при больших x. При положительных значениях проявляются более плавные изменения.
